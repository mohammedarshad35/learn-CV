```python
#   MATPLOTLIB reads images as RGB --> RED GREEN BLUE in this order
#   where as OpenCV reads as BGR ----> BLUE GREEN RED in this order
# in order to resolve this we need to transform.
```

```python
trans_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(trans_img)
```

```python
img_gray=cv2.imread(path+'/IMG_0402.JPG',cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray,cmap='gray')    # here matplotlib read img inn different color format so we need to use cmap ='gray' again inoorder to see gray color
```

## Resizing images

```python
plt.imshow(trans_img)
```

```python
new_img=cv2.resize(trans_img,(800,700))
```

```python
plt.imshow(new_img)
```

```python
trans_img.shape
```

```python
w_ratio=0.5
h_ratio=0.5
```

```python
new_img=cv2.resize(trans_img,(0,0),trans_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
trans_img.shape
```

```python
new_img.shape
```

```python
new_img=cv2.flip(trans_img,0)
```

```python
plt.imshow(new_img)
```

```python
cv2.imwrite('new_pic.jpg',new_img)
```
